#### Vraag 2h
Het elektrische veld een afstand $z$ boven het midden van een uniform geladen schijf met een straal $R$ en oppervlakte lading $\sigma$ wordt gegeven door:
$$ \mathbf{E} = \frac{\sigma}{2\epsilon_0} \bigg[ 1- \frac{z}{\sqrt{R^2 + z^2}}\bigg] \hat{\mathbf{z}}$$
Zie probleem 2.6 in het boek van Griffith.

1. Wat moet de hoogte $z$ zijn boven de schijf zodat het elektrische veld door de schijf 10% is in vergelijking met dat van een oneindig geladen plaat?

Voeg hieronder je analytische uitwerking toe.

##### Antwoord 2h

Voor een oneindige plaat geldt:
$E_{\infty} = \frac{\sigma}{2\epsilon_0}$

Voor een schijf geldt:
$E(z) = \frac{\sigma}{2\epsilon_0}\left(1 - \frac{z}{\sqrt{R^2 + z^2}}\right)$

We willen dat het veld van de schijf 10% is van dat van de oneindige plaat:

$E(z) = 0.10 \, E_{\infty}$

Invullen geeft:

$\frac{\sigma}{2\epsilon_0}\left(1 - \frac{z}{\sqrt{R^2 + z^2}}\right)
= 0.10 \, \frac{\sigma}{2\epsilon_0}$

We delen beide kanten door $\frac{\sigma}{2\epsilon_0}$:

$1 - \frac{z}{\sqrt{R^2 + z^2}} = 0.10$

Dus:

$\frac{z}{\sqrt{R^2 + z^2}} = 0.90$

Nu kwadrateren we beide kanten:

$\frac{z^2}{R^2 + z^2} = 0.81$

Vermenigvuldigen met $R^2 + z^2$:

$z^2 = 0.81(R^2 + z^2)$

Uitwerken:

$z^2 = 0.81R^2 + 0.81z^2$

Alles naar één kant:

$z^2 - 0.81z^2 = 0.81R^2$

$0.19z^2 = 0.81R^2$

Dus:

$z^2 = \frac{0.81}{0.19} R^2$

$z^2 \approx 4.26 R^2$

Daaruit volgt:

$z \approx 2.06 R$

De hoogte moet dus ongeveer $z \approx 2.1R$ zijn zodat het veld nog maar 10% is van dat van een oneindig geladen plaat.

#### Vraag 3b: De oppervlakte lading in het midden van een geleidende schijf op constant potentiaal

1. Met welke formule kan je de oppervlaktelading benaderen in het midden van de schijf($r=0$, $z=0$)?
2. Print netjes de benadering van de oppervlakte lading in het midden van de schijf.

##### Antwoord 3b.1

De oppervlakteladingsdichtheid op een geleider kan worden bepaald met de randvoorwaarde voor het elektrische veld net buiten het oppervlak:

$\sigma = \epsilon_0 E_\perp$

Hierbij is $E_\perp$ de component van het elektrische veld loodrecht op het oppervlak.  

In het midden van de schijf ($r = 0$, $z = 0$) is dit de z-component van het veld net boven de schijf:

$\sigma \approx \epsilon_0 E_z(0^+)$

#### Vraag 3c: De oppervlaktelading van een geleidende schijf op constant potentiaal

De oppervlakte lading $\sigma(r)$ van een geleidende schijf met straal $R$ is niet uniform verdeeld over de schijf, maar wordt gegeven door (Problems 2.58 and 2.63 Griffiths 5th edition):
$$\sigma(r) = \frac{Q}{2 \pi R \sqrt{R^2 - r^2}}$$

1. Bereken de totale lading op de schijf met behulp van vraag 3a en 3b.
2. Vergelijk de oppervlakte lading in het centrum van de geleidende schijf met dat van een uniform geladen schijf met dezelfde totale lading.
3. Plot de oppervlakte lading van de geleidende schijf met de berekende lading $Q$, tussen $r = 0$ en $R$.
4. Plot in hetzelfde figuur de oppervlaktelading van een uniform geladen schijf met dezelfde lading $Q$.


````python
# 1

idx_r_zero = np.argmin(np.abs(r_for_E))
idx_z_near_zero = np.argmin(np.abs(z_for_E))

Ez_center = Ez[idx_z_near_zero, idx_r_zero]
sigma_center = epsilon0 * Ez_center

Q_total = 2 * np.pi * disk_radius**2 * sigma_center

print(f"Totale lading op de geleidende schijf Q = {Q_total:.3e} C")

# 2

sigma_uniform = Q_total / (np.pi * disk_radius**2)

print(f"Oppervlaktelading in centrum (geleidend) = {sigma_center:.3e} C/m^2")
print(f"Oppervlaktelading uniforme schijf = {sigma_uniform:.3e} C/m^2")

# 3

r_plot = np.linspace(0, disk_radius * 0.999, 500)
sigma_conducting = Q_total / (2 * np.pi * disk_radius * np.sqrt(disk_radius**2 - r_plot**2))

# 4

sigma_uniform_plot = np.ones_like(r_plot) * sigma_uniform

plt.figure()
plt.title("Oppervlaktelading σ(r)")
plt.plot(r_plot, sigma_conducting, label="Geleidende schijf")
plt.plot(r_plot, sigma_uniform_plot, label="Uniform geladen schijf")
plt.xlabel("r (m)")
plt.ylabel("σ (C/m^2)")
plt.legend()
plt.show()
````
