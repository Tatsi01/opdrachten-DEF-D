```python
# Schrijf hier een python functie die de capaciteit van een plaatcondensator uitrekent. Zorg dat de functie
# als input de ontwerpparameters van de vlakke plaat formule accepteert (oppervlakte, afstand en diëlektrische
# constante) en als output de capaciteit van de vlakke plaat condensator met die ontwerpparameters.
# Je gaat deze formule de rest van de dag gebruiken: controleer hem goed!
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
def Capaciteit(d,relPerm,A):
    output = (relPerm*8.854e-12*A)/(d) 
    print(output)
    return output*(10**12)

# Optioneel: schrijf je functie zodanig dat je arrays als input kan geven en (dus) ook arrays als output.
```

```python
# Gebruik je formule om uit te rekenen wat je verwacht dat de capaciteit van jullie vlakke plaat condensatoren
# gaat zijn.

C = [Capaciteit(2.15e-3,3.3,0.01)]


print(f"Berekende capaciteit plaatcondensator groep 50 is {C[0]:.1f} pF")

```

```python
# Plot hier de verwachte ontwerpgrafiek van jullie sensoren.

VeerConst = 500
#d = np.linspace(10e-3,0,100)
F = 9.81 * np.linspace(500e-3,100e-3)
print(17e-3 - F/VeerConst)

plt.figure()
plt.title('verwachtte verband')
plt.plot(F,Capaciteit((17e-3 - F/VeerConst),1,0.01),label = 'verwachtte lijn')
plt.xlabel('F(N)')
plt.ylabel('C(pF)')
plt.legend()
plt.show()
```

## Opdracht 7: Plot de kalibratiegrafiek.

```python
# Plot hier je gemeten kalibratiegrafiek en plot in dezelfde figuur ook je verwachte ontwerpgrafiek.

Force = 9.81*np.array([467e-3,390e-3,356e-3,340e-3,317.5e-3,305e-3,268e-3])
cap = np.array([760,730,700,680,645,615,590]) 
yerror = np.array([30,30,30,30,30,30,30])
xerror= np.array([0.05e-3,0.05e-3,0.05e-3,0.05e-3,0.05e-3,0.05e-3,0.05e-3]) * 9.81

values, cov = curve_fit(Capaciteit,(17e-3 - Force/VeerConst),cap)
plt.title('Metingen en verwachtte lijn')
plt.errorbar(Force,cap,fmt ='o',xerr=xerror,yerr = yerror,label = 'metingen')
plt.plot(F,Capaciteit((17e-3 - F/VeerConst),1,0.01),label = 'verwachtte metingen')
plt.plot(Force,Capaciteit((17e-3 - Force/VeerConst),values[0],values[1]), label ='y = (8.854*(10^-12) * ' + str(round(values[0],3)) +'*'+ str(round(values[1],3)) + ')/d')
plt.xlabel('F(N)')
plt.ylabel('C(pF)')
plt.legend()
plt.show()
```
